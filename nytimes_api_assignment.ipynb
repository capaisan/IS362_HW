## Working with the New York Times API

This notebook demonstrates how to fetch data from the NY Times Most Popular API and convert it to a pandas DataFrame.

```python
import requests
import pandas as pd
import json
```

```python
API_KEY = 'EaR5WKpqdTDAafWXVx9Fm4mWnt8QrhNk'

# Most Popular API endpoint - get most viewed articles from last 7 days
url = f'https://api.nytimes.com/svc/mostpopular/v2/viewed/7.json?api-key={API_KEY}'
```

## Fetch Data from API

```python
# Make the API request
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    data = response.json()
    print(f"Successfully fetched {data['num_results']} articles")
else:
    print(f"Error: {response.status_code}")
```

```python
# Look at the structure of one article
print(json.dumps(data['results'][0], indent=2))
```

## Convert to DataFrame

```python
# Extract the articles from the results
articles = data['results']

# Create DataFrame
df = pd.DataFrame(articles)

# Display basic info
print(f"Shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
```
