##We import pandas and use the mysql connector to load in the mysql database and use its contents for the data analysis. 


```python
import pandas
import mysql.connector

cnx = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Diagnoses-Curable-Diner",
    database="project1",
    port="80"
)

query = "select * from flight_delays"

data = pandas.read_sql(query, cnx)

cnx.close()

print(data)
```

###Here we comparing the two airlines delay rate 

```python
stats = data.groupby(['airline', 'status'])['count'].sum().unstack()
stats['Total flights'] = stats['on time'] + stats['delayed']
stats['Delay Rate %'] = (stats['delayed'] / stats['Total flights']) * 100

print('Total Delay Rate:')
print(stats)
```

###Here we comparing the those same airlines, but determining the percentage that a flight was delayed based on the destinations

```python
destination_stats = data.pivot_table(
    index=['airline', 'destination'],
    columns='status',
    values='count'
).reset_index()

destination_stats['Total flights'] = destination_stats['on time'] + destination_stats['delayed']
destination_stats['Delay Rate %'] = (destination_stats['delayed'] / destination_stats['Total flights']) * 100

print("Delay Rate by Destination:")
print(destination_stats.to_string(index=False))

```
